```python
string = input("Enter a string: ")
count = {}
for a in string:
    if a in count:
        count[a] += 1
    else:
        count[a] = 1

max = max(count, key=count.get)
count = count[max]

print(f"The highest occurring character is '{max}' with a count of {count}.")
```
